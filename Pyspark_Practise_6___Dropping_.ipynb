```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder.getOrCreate()
```

## Question 8

```python
df = spark.read.csv('wine.csv')
```

```python
df.show()
```

```python
def fun(x , df):
  df = df.drop(x)
  return df

ind = [0,3,6,8,11,12,13]
col = df.columns
for i in ind :
  df = fun(col[i], df)

df.show()
```

```python
mns = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
col = df.columns

for i in range(len(col)):
  df = df.withColumnRenamed(col[i] , mns[i])

df.show()
```
